### Euclidean distance 
Euclidean distance is the most common use of distance. 
It examines the  root of  square difference between coordinates  of a pair of objects


```python
from scipy.spatial import distance
```

```python
a = (0,3,4,5)
b = (7,6,3,-1)
dst = distance.euclidean(a, b)
print(dst)
```

### Manual Calculation


## Manhattan distance

It represents the distance  between  point in  a city road grid
It examines  the absolute difference between the coordinates of a  pair of objects


```python
a = (0,3,4,5)
b = (7,6,3,-1)
dst = distance.cityblock(a,b)
print(dst)
```

Manual Calculation :

= (0-7)+(3-6)+(4-3)+(5+1)
= 7+3+1+6
=17

## Chebyshev distance
It is maximum value distance
It examines  the absolute  magnitude of  the difference  between coordinates of pairs of objects


```python
a = (0,3,4,5)
b = (7,6,3,-1)
dst = distance.chebyshev(a,b)
print(dst)
```

## Manual calculation :
= max{|0-7|,|3-6|,|4-3|,|5+1|}

= max{7,3,1,6}

= 7

## Minkowski Distance

Generalized metric distance
When Lamda = 1, city block distance
When lamda = 2, it become euclidean
When lamda > 2 , it is Chebyshev distance


```python
a = (0,3,4,5)
b = (7,6,3,-1)
p = 3 # where p is the root value
dst = distance.minkowski(a,b,p)
print(dst)
```

## Jaccard Distance

Jaccard distance, is a measure of howdissimilartwo sets
 
 


```python
# Jaccard similarty
import numpy as np
from sklearn.metrics import jaccard_similarity_score
a = [0,2,5]
b = [0,2,3]
out = jaccard_similarity_score(a, b)
out * 100
```

```python
## Jaccard dissimilarity
100 - (out * 100)
```
